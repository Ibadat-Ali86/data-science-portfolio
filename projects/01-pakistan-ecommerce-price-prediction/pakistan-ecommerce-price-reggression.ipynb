```python
import pandas as pd
import os
import kagglehub

# Download dataset directly from Kaggle
print("Downloading dataset from Kaggle...")
path = kagglehub.dataset_download("zusmani/pakistans-largest-ecommerce-dataset")
print(f"Dataset downloaded to: {path}")

# List to hold dataframes
df_list = []

# Loop through the downloaded directory and read all CSV files
for dirname, _, filenames in os.walk(path):
    for filename in filenames:
        if filename.endswith('.csv'):
            file_path = os.path.join(dirname, filename)
            print(f"\nLoading: {filename}")
            
            # Reading with low_memory=False is often necessary for this 
            # specific dataset due to mixed data types in some columns
            temp_df = pd.read_csv(file_path, low_memory=False)
            df_list.append(temp_df)
            print(f"  ✓ Loaded {len(temp_df):,} rows")

# Check if any files were found
if len(df_list) == 0:
    raise ValueError("No CSV files found in downloaded dataset!")

# Concatenate all dataframes into one
df = pd.concat(df_list, ignore_index=True)
print(f"\n✓ Successfully loaded {len(df):,} total rows!")

```

## **Outlier Detection & Handling**

- Before feature engineering, let's detect and handle outliers in our target variable (unit_price).
- Although we are using log scaling that is also a technique for outlier handling but for more clarifications i am going to do outlier analysis also!

```python
# Check for outliers using IQR method
Q1 = df['unit_price'].quantile(0.25)
Q3 = df['unit_price'].quantile(0.75)
IQR = Q3 - Q1

lower_bound = Q1 - 1.5 * IQR
upper_bound = Q3 + 1.5 * IQR

print("OUTLIER DETECTION")
print("-" * 40)
print(f"Q1 (25th percentile): PKR {Q1:,.2f}")
print(f"Q3 (75th percentile): PKR {Q3:,.2f}")
print(f"IQR: PKR {IQR:,.2f}")
print(f"Lower Bound: PKR {lower_bound:,.2f}")
print(f"Upper Bound: PKR {upper_bound:,.2f}")

# Count outliers
outliers = df[(df['unit_price'] < lower_bound) | (df['unit_price'] > upper_bound)]
print(f"\nOutliers found: {len(outliers):,} ({len(outliers)/len(df)*100:.1f}%)")
```

```python
# Visualize outliers
plt.figure(figsize=(10, 4))

plt.subplot(1, 2, 1)
plt.boxplot(df['unit_price'].dropna())
plt.ylabel('Unit Price (PKR)')
plt.title('Before Outlier Handling')
plt.yscale('log')

plt.subplot(1, 2, 2)
df['unit_price'].hist(bins=50, color='skyblue', edgecolor='black')
plt.axvline(upper_bound, color='red', linestyle='--', label=f'Upper: {upper_bound:,.0f}')
plt.xlabel('Unit Price (PKR)')
plt.title('Price Distribution')
plt.xlim(0, 50000)
plt.legend()

plt.tight_layout()
plt.savefig('outlier_analysis.png', dpi=300, bbox_inches='tight')
plt.show()
```

```python
# Handling outliers - Remove only EXTREME outliers (beyond 99.5th percentile)
# We keep most outliers because they are valid expensive products

extreme_upper = df['unit_price'].quantile(0.995)  # 99.5th percentile
extreme_lower = 0  # Remove negative or zero prices

print("OUTLIER HANDLING")
print("-" * 40)
print(f"Removing prices > PKR {extreme_upper:,.2f} (99.5th percentile)")
print(f"Removing prices <= PKR {extreme_lower}")

# Count before
before_count = len(df)

# Remove extreme outliers only
df = df[(df['unit_price'] > extreme_lower) & (df['unit_price'] <= extreme_upper)]

# Count after
after_count = len(df)
removed = before_count - after_count

print(f"\nRecords before: {before_count:,}")
print(f"Records after: {after_count:,}")
print(f"Removed: {removed:,} ({removed/before_count*100:.2f}%)")
```

**Outlier Handling Approach:**
- We only removed EXTREME outliers (top 0.5% prices)
- Most "outliers" are kept because they are valid expensive products
- Removing all outliers would hurt our model's ability to predict high prices
- This approach removes only truly anomalous data points

## **Features & Target**

```python
#  Target (log-price)
Y = np.log1p(df["unit_price"])

#  Numeric features
numeric_features = [
    "quantity_ordered",
    "order_total_amount",
    "discount_value",
    "discount_rate",
    "customer_tenure_days",
    "order_year",
    "order_month",
    "order_dayofweek",
    "is_weekend",
    "is_bulk_order"
]

# Categorical features
categorical_features = [
    "product_category",
    "payment_type",
    "order_status"
]

X = df[numeric_features + categorical_features]
X.head()
```

```python
#splitting the dataset into 15% validation,15% testing and 70% for training
X_temp,X_test,y_temp,y_test = train_test_split(X,Y,test_size = 0.15, random_state = 42) #testing data from full dataset

X_train,X_val,y_train,y_val = train_test_split(X,Y,test_size = 0.1765, random_state = 42) #just getting the 15% for validation using remaining 85% of data
                                                                                           #then using the 70% in training
print("Train size:", X_train.shape)
print("Validation size:", X_val.shape)
print("Test size:", X_test.shape)
```

This code carefully partitions the dataset to enforce proper machine learning evaluation discipline.

The training set is used to learn model parameters.

The validation set is used to compare models and tune hyperparameters without bias.

The test set remains untouched until the very end, providing an unbiased estimate of real-world performance.

By splitting the test set first and deriving the validation fraction mathematically, the code avoids data leakage and produces statistically honest performance estimates, assuming IID data.

## **Preprocessing + Model in a Pipeline**

We will use these techniques:

- OneHotEncoder for categorical columns

- Pass‑through for numeric columns

- RandomForestRegressor as the model (good default, no scaling needed)



```python
#appliying preprocessing on numeric and categorical cols and then saving their outputs in final tabular format
numeric_transformer = 'passthrough' #keep it unchanged

#applying onehotencoding on categorcal cols 
categorical_transformer = OneHotEncoder(handle_unknown = 'ignore', sparse_output = False)

#making final tabular matrix
preprocessor = ColumnTransformer(transformers = [('num',numeric_transformer,numeric_features), ('cat',categorical_transformer,categorical_features)])
```

- This code builds a preprocessing layer that converts raw mixed-type tabular data into a clean numeric feature matrix suitable for a RandomForest model.

- Numeric features are passed unchanged because the model is scale-invariant.

- Categorical features are safely one-hot encoded with protection against unseen categories.

## **Ensemble Model 1 – Random Forest (Baseline Ensemble)**

Why Random Forest?

- Handles non‑linear relationships.

- Works well with mixed feature types.

- Good robust baseline ensemble.

```python
def print_regression_metrics(y_true_log, y_pred_log, label="Model"):
 
    mae_log = mean_absolute_error(y_true_log, y_pred_log)
    rmse_log = np.sqrt(mean_squared_error(y_true_log, y_pred_log))
    r2 = r2_score(y_true_log, y_pred_log)

    # back-transform to original price
    y_true = np.expm1(y_true_log)
    y_pred = np.expm1(y_pred_log)
    mape = np.mean(np.abs((y_true - y_pred) / y_true)) * 100

    print(f"=== {label} ===")
    print(f"MAE (log):  {mae_log:.4f}")
    print(f"RMSE (log): {rmse_log:.4f}")
    print(f"R²:         {r2:.4f}")
    print(f"MAPE (PKR): {mape:.2f}%\n")
```

```python
# applying bagging
rf = RandomForestRegressor(n_estimators = 200, max_depth = None,min_samples_split = 2,min_samples_leaf = 1,random_state = 42, n_jobs = -1)

#now making pipline that will automatically take the raw data preprocess it and then train the model
model = Pipeline(steps = [('preprocess',preprocessor),('rf',rf)])

model.fit(X_train,y_train)

#Evaluation on validation
y_pred = model.predict(X_val)
print_regression_metrics(y_val,y_pred,label = 'Random Forest (baseline)') #Computing MAE,RMSE & R squared error
```

**Interpretation:**
- **R² = 0.92**: Model explains 92% of variance - excellent performance
- **MAE (log) = 0.25**: Low average error on log scale
- **RMSE (log) = 0.50**: Acceptable deviation
- **MAPE = 66%**: High because of small price items where small errors = big percentages

Random Forest performs very well overall.

- This code creates a single, end-to-end machine learning system that automatically prepares data and trains a Random Forest model in a safe, repeatable way.

- Random Forest handles complex patterns without scaling

- Preprocessing is always applied correctly

- Training and prediction use the exact same logic

## **Ensemble Model 2 – Gradient Boosting (Advanced Ensemble)**

```python
#applying the Boosting techniques
gb_model = GradientBoostingRegressor(
    n_estimators = 300,
    learning_rate = 0.05,
    max_depth = 3, 
    random_state = 42
)

gb_pipeline = Pipeline(steps = [('preprocess',preprocessor),('gb',gb_model)])

gb_pipeline.fit(X_train,y_train) 

gb_predict = gb_pipeline.predict(X_val)

print_regression_metrics(y_val, gb_predict, label="Gradient Boosting (baseline)")
```

**Interpretation:**
- **R² = 0.85**: Model explains 85% of variance - good but worse than Random Forest
- **MAE (log) = 0.48**: Higher error than Random Forest
- **RMSE (log) = 0.70**: Higher deviation than Random Forest
- **MAPE = 114%**: Very high error percentage

Gradient Boosting baseline is weaker than Random Forest. Needs hyperparameter tuning to improve.

## **Hyper Parameters Tunning on Gradient Boosting**


Hyperparameter tuning is critical for Gradient Boosting because small changes can drastically affect performance.

- Using RandomizedSearchCV + CV:

- Systematically explores combinations

- Prevents overfitting to training data

- Keeps computation manageable

```python
#this is neccessary bcz small changes can improve the performance. so for this we have to find 
# optimal estimator or pipeline and then have to train model on that
para_db = {
    'gb__n_estimators': [200,300,400],
    'gb__learning_rate': [0.01,0.03,0.05,0.1],
    'gb__max_depth' : [2,3,4],
    'gb__subsample': [0.8,1.0]# Fraction of data used to train each tree.
}

#finding the best commbinations
gb_random_srch = RandomizedSearchCV(estimator = gb_pipeline,param_distributions = para_db,n_iter = 10,cv = 3, scoring = 'neg_mean_absolute_error',n_jobs = -1,random_state= 42,verbose = 1)

gb_random_srch.fit(X_train,y_train) #model will be trained on cv for best combinations
print("best params: ",gb_random_srch.best_params_)
gb_best = gb_random_srch.best_estimator_

#evaluating on validation 
gb_predict_best = gb_best.predict(X_val)

print_regression_metrics(y_val, gb_predict_best, label="Gradient Boosting (tuned)")
```

## **Choose Best Ensemble & Evaluate on Test Set**

```python
print("Comparing models on VALIDATION:")
print_regression_metrics(y_val, y_pred, label="Random Forest (baseline)")
print_regression_metrics(y_val, gb_predict_best, label="Gradient Boosting (tuned)")

# Evaluating on the TEST data
y_testing_pred = gb_best.predict(X_test)
print_regression_metrics(y_test, y_testing_pred, label="Gradient Boosting (tuned) - TEST")
```

## **Model Performance by Price Range**

Let's check how well our model predicts prices for cheap vs expensive products.


```python
# Geting predictions from Random Forest model
y_val_pred_rf = model.predict(X_val)

# Converting back from log scale to actual prices
actual_prices = np.expm1(y_val)
predicted_prices = np.expm1(y_val_pred_rf)

# Creating simple price categories
price_category = pd.cut(actual_prices, 
                        bins=[0, 2000, 10000, float('inf')],
                        labels=['Low (0-2K)', 'Medium (2K-10K)', 'High (10K+)'])

# Checking R² score for each category
print("MODEL PERFORMANCE BY PRICE RANGE")
print("-" * 40)

for category in ['Low (0-2K)', 'Medium (2K-10K)', 'High (10K+)']:
    mask = price_category == category
    if mask.sum() > 0:
        r2 = r2_score(actual_prices[mask], predicted_prices[mask])
        count = mask.sum()
        print(f"{category}: R² = {r2:.3f} (n={count:,})")
```

**Surprising Finding from Price Range Analysis:**

| Price Range | R² Score | Performance |
|-------------|----------|-------------|
| Low (0-2K) | -3.26 | Very Poor |
| Medium (2K-10K) | 0.18 | Poor |
| High (10K+) | 0.90 | Excellent |

**Why does this happen?**
- Low prices have high variance (many different cheap products)
- High prices are more predictable (premium products follow patterns)
- Overall R²=0.92 is driven by high-price items

```python
#  Actual vs Predicted for each price range
plt.figure(figsize=(12, 4))

for i, category in enumerate(['Low (0-2K)', 'Medium (2K-10K)', 'High (10K+)']):
    plt.subplot(1, 3, i+1)
    mask = price_category == category
    plt.scatter(actual_prices[mask], predicted_prices[mask], alpha=0.3, s=5)
    
    # Add perfect prediction line
    max_val = max(actual_prices[mask].max(), predicted_prices[mask].max())
    plt.plot([0, max_val], [0, max_val], 'r--', label='Perfect Prediction')
    
    plt.xlabel('Actual Price (PKR)')
    plt.ylabel('Predicted Price (PKR)')
    plt.title(f'{category}')
    plt.legend()

plt.tight_layout()
plt.savefig('price_range_performance.png', dpi=300, bbox_inches='tight')
plt.show()
```

**Key Finding:**
- Model performs BEST for high prices (R² = 0.90)
- Model struggles with low prices (R² = -3.26)
- This is opposite to what we initially expected!
- Reason: High-value products have more predictable pricing patterns
